```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
df.columns
```

```python
# Task 1: Basic Data Exploration
num_rows, num_columns = df.shape
data_types = df.dtypes
missing_values = df.isnull().sum()

print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
print("\nData types of each column:")
print(data_types)
print("\nMissing values in each column:")
print(missing_values)

```

```python
# Task 2: Descriptive Statistics
salary_column = 'TotalPay'
basic_stats = df[salary_column].describe()
salary_range = df[salary_column].max() - df[salary_column].min()
salary_std = df[salary_column].std()

print("\nBasic Statistics:")
print(basic_stats)
print(f"\nRange of Salaries: {salary_range}")
print(f"Standard Deviation of Salaries: {salary_std}")

```

```python
# Task 3: Data Cleaning
# Handle missing data by filling with the mean salary
df[salary_column].fillna(df[salary_column].mean(), inplace=True)
```

```python
# Task 4: Basic Data Visualization
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Importing the matplotlib module

# Load the dataset
df = pd.read_csv('Salaries.csv')

# ... (rest of the code remains unchanged)

plt.figure(figsize=(10, 6))
plt.hist(df[salary_column], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

department_counts = df['JobTitle'].value_counts().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Job Titles Distribution')
plt.show()
```

```python
# Task 5: Grouped Analysis
grouped_stats = df.groupby('Year')[salary_column].agg(['mean', 'median', 'min', 'max'])
print("\nGrouped Analysis:")
print(grouped_stats)
```

```python
# Task 6: Simple Correlation Analysis
correlation = df[salary_column].corr(df['BasePay'])
plt.figure(figsize=(8, 6))
plt.scatter(df['BasePay'], df[salary_column], color='green', alpha=0.7)
plt.title('Scatter Plot: Salary vs Base Pay')
plt.xlabel('Base Pay')
plt.ylabel('Salary')
plt.show()
print("\nCorrelation between Salary and Base Pay:", correlation)
```
